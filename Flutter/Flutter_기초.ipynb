변수의 선언은 타입과 함께 이뤄진다.

물론 타입을 정의하지 않아도 되지만, 추후 코드 관리나 협업을 위해 타입을 함께 선언하는 것이 좋다.

변수는 크게 **불변형 변수**와 **가변형 변수**로 구성되어 있다.

### **2. 불변형 변수**

```
void main() {

  // bool (참/거짓)
  bool isTrue = true;

  // 정수형의 int
  int num = 100;

  // 실수형의 double
  double num2 = 3.14;
  double num3 = 3;

  // 문자열형 String
  String string = 'Hellow World';

  // Null형 null
  Null thisIsNull = null;

  // 위에 선언한 변수 사용하기
  print(isTrue);      // true
  print(num);         // 100
  print(num2);        // 3.14
  print(num3);        // 3
  print(string);      // Hellow World
  print(thisIsNull);  // null
}
```

불변형 변수는 **bool, int, double, String, Null** 의 5가지로 구성된다.

Type을 먼저 선언하고, 바로 변수명이 온다.

단순히 선언만 한다고 사용될 수 없기 때문에 해당 type에 대한 값이 함께 선언된다.

중요한것은 마지막에 세미콜론( ; )이 함께 작성되어야 실행이 가능하다.


### **3. 가변형 변수**

image.png

가변형 변수는 var, dynamic 2가지 유형이 있다.

var과 dynamic에서 차이는 타입 변경이 가능한가이다.

- **var** : 타입 변경이 불가능
- **dynamic** : 중간에 타입 변경 가능

**var**

위에 var 예시를 보면, 1로 선언을 해서 정수로 초기화가되면

value = '문자'로 변경하려해도 변경되지 않고 error message가 나타난다.

변경 가능한 것은 같은 타입 안에서 가능하다.

value = 2 ; 로 변경했을 때 변경이 이뤄지는 것을 확인할 수 있다.

**dynamic**

dynamic의 경우는 중간에 타입 변경이 가능하다.

위의 예시처럼 100인 정수형으로 초기화되면, var는 변경이 불가했는데

'Hello'라는 문자열로 변경이 가능한 것을 확인할 수 있다.

### **산술 연산자**



- case1

  ```
  void main() {
    // 사칙 연산 (+, -, *, /)
    int num1 = 10;
    int num2 = 20;

    print(num1 + num2); // 30
    print(num1 - num2); // -10
    print(num1 * num2); // 200
    print(num1 / num2); // 0.5
    }
  ```

정수 타입으로 먼저 선언을 하고 값을 위에 할당한 값으로 연산을 통해 정의할 수도 있다.

- case2

  ```
  void main() {
    // 사칙 연산 (+, -, *, /)
    int num1 = 10;
    int num2 = 20;

    print(num1 + num2);   // 30

    int num3;
    num3 = num1 + num2;
    print(num3);          // 30
    }
  ```

다만 나눗셈은 주의해야 하는데, case1에서는 나눗셈시 0.5의 결과를 나타냈지만 (상속),

case2에서는 타입이 정의되어 있어 error가 발생된다. (불가)

- error 예시
  ```
  void main() {
    // 사칙 연산 (+, -, *, /)
    int num1 = 10;
    int num2 = 20;

    print(num1 + num2);   // 30

    int num3;
    num3 = num1 / num2;   // error
    print(num3);          // error
    }
  ```

따라서 double로 바꿔주거나, var, dynamic을 활용할 수 있다.

**2. Null Safety 개념 도입**

Dart 2.12 버전부터 Null값의 안정성을 위해 Null Safety 개념이 도입되었다.

Nullable Type과 Non-nullable Type으로 구성되며 변수 뒤에 물음표나 느낌표가 붙음으로서 구분되어 사용된다.

**1) Nullable Type (null 허용)** : **?**
  - 변수 뒤에 ? 붙여 활용

**2) Non-nullable Type (null 비허용)** : **!**
  - 변수 뒤에 ! 붙여 활용
