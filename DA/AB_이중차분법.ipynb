### 독립 변수 vs 종속 변수

- 독립 변수 : 다른 변수에 영향을 받지 않으며 종속 변수 변화에 대해 부분적으로 설명할 수 있는 요소
  - 변화 요소
  - 부분적으로 설명 : 독립변수와 종속변수가 1:1의 인과관계를 갖고 있다 말하기 어려움
  - 상관성이 높은 독립 변수를 찾아야
- 종속 변수 : 독립 변수에 영향을 받아 변화하는 변수
  - 측정이 가능해야
  - 변화량에 따라 실험의 성과를 판별할 수 있음
  - 어떤 기준으로 측정할 것인지 명확한 사전 기준 필요

## A/B Test

- 실무자가 세운 가설이 현실에서 유의미하다는 보장이 없기 때문에 진행

- A와 B를 대조해 가설을 검정하는 방법
  - 종속변수에 영향을 주는 독립 변수를 찾기 위한 노력

- 실무에서 데이터 기반 의사 결정 하지 않으면 더욱 큰 비용이 발생하거나, 치명적인 결과가 발생될 수 있어 진행 필요

- 사례
  - 가설1 : 배너 위치를 변경하면 CTR(클릭전환율)이 증가할 것이다.
  - 가설2 : 장바구니 팝업 문구를 변경하면 CVR(주문 전환율)이 증가할 것이다
  - 가설3 : 5% 할인 쿠폰보다 5천원 할인쿠폰이 주문이 더 많이 일어날 것이다
    - 독립 변수 : 배너 위치, 팝업 문구, 할인 쿠폰
    - 종속 변수 : CTR(클릭전환율), CVR(주문전환율), 주문

- 단순히 가설에 대한 실험이 유의미하다는 보장이 없음
  - 샘플 사이즈나 실험 기간 계산 등 통계 수학을 적용
  - 통제 변수를 컨트롤해 신뢰성 높여야
- 경우에 따라 Test 비용이 많이 든다

### 이중차분법 (Difference In Differences)

image.png

- 이중차분법 : (A-a) - (B-b)
  - 실험군의 처치 이후 지표와 처치 전 지표 차이를 대조군의 같은 시점 전후 지표 차와 한 번 더 차이 주는 것
  - 실험자가 지정한 독립변수 이외의 변수들의 개입을 최대한 제외
  - 의도적으로 준 독립 변수의 효과만을 확인할 수 있게 됨

- 독립변수와 종속변수간 인과 효과를 측정하기 위한 인과추론법
- 대조군과 실험군의 두 그룹의 처치 시점 전후로 지표를 이중으로 처분하는 것
  - control군의 처치 전후 지표와 treament군 전후 지표
  - 시계열 데이터를 비교
  - 어떠한 처치 없이도 나타날 결과를 결과 지표에서 제외하는 방법

- 실험군과 대조군
  - 실험군(Treatment Group) : 실험자가 종속변수와 상관성이 높아보이는 독립변수에 변화를 주는 그룹
  - 대조군(Control Group) : 실험군과 비교하기 위해 처치를 하지 않는 그룹
    - 실험군과 비교하기 위해 다른 변수는 동일하게 통제
    - 처치는 하지 않음

- 이중차분법 위한 그룹별 주문수 집계
  - 그룹별, 기간별 order를 count
- 이탈 고객이 얼마나 많이 주문했는지가 중요한 것이 아니라, 이탈 고객이 처치 후 얼마나 복귀했는지가 중요
  - 한번이라도 주문하면 1로 count
  - 처치전 맴버 1명이 5번 주문해도 1로 count

````python
df2 = df[['mem_no','group','ord_no','ord_dt']]
df2.head()
````

- ord_dt에서 2023-06-05, 2023-06-07은 모두 처치 전일
  - 처치 전과 후로 분류

````python
df2['ord_dt'] = pd.to_datetime(df2['ord_dt'])
````

````python
df2.info()
````

````python
import warnings
warnings.filterwarnings('ignore')
````

````python
처치전 = (df2['ord_dt'].between('2023-06-05','2023-06-11'))
처치후 = (df2['ord_dt'].between('2023-06-12','2023-06-18'))

df2['period'] = np.select([처치전, 처치후],['처치전','처치후'], default=np.nan)
df2['period'] = df2['period'].replace('nan', np.NaN)
df2.head()
````

````python
df2.period.isna().sum()
````

- 그룹별, peroid 별로 mem_no 중복 제거하고 count

````python
df3 = df2.groupby(['group','period'])['mem_no'].nunique().reset_index(name='ORD_CNT').dropna()
df3
````

- SQL로 진행되는 경우

  - Table1
  ```
  SELECT ctt.mem_no,
            "group",
             ord_no,
            CASE WHEN ord_dt BETWEEN '2023-06-05' and '2023-06-11' then '처치전'
                 WEHN ord_dt BETWEEN '2023-06-12' and '2023-06-18' then '처치후' end as period
  FROM coupon_target_table ctt
  LEFT JOIN order_master_did omd ON ctt.mem_no = omd.mem_no
  ```
  
  - Table1을 통해 집계 (group, 기간 별 order건 count)
  ```
  WITH
  T1 as (
      SELECT ctt.mem_no,
                "group",
                ord_no,
                CASE WHEN ord_dt BETWEEN '2023-06-05' and '2023-06-11' then '처치전'
                    WEHN ord_dt BETWEEN '2023-06-12' and '2023-06-18' then '처치후' end as period
      FROM coupon_target_table ctt
      LEFT JOIN order_master_did omd ON ctt.mem_no = omd.mem_no
  )
  # 그룹별로 period별로 mem-no를 중복 제거하고 count
      SELECT "group",
              period,
              count(DISTINCT mem_no) AS ord_cnt
      FROM T1
      WHERE period IS NOT NULL
      GROUP BY 1,2
  ```
